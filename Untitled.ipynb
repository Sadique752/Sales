```python
premier_league.query("HomeTeam in ['Arsenal', 'Chelsea', 'Everton']").pivot_table(
                           index="HomeTeam",
                           columns="season",
                           values="HomeGoals",
                           aggfunc="mean",
                           margins=True).style.background_gradient(cmap="RdYlGn", axis=None)
```
